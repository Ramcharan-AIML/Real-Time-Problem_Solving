# EXPENSE TRACKER MINI - REAL WORLD PROBLEM SOLVING

```python
# OUTPUT SHOULD BE LIKE THIS

"""

1. Add Expense
2. View Expenses
3. Show Total
4. Exit

Enter choice: 1
Enter item: Tea
Enter amount: 20
Expense added successfully

Enter choice: 2
Tea - 20

Enter choice: 3
Total spent: 20

"""
```

```python
items_lst = []

def add_expenses(name , amount):

    Items = {
        "Item" : name,
        "Amount": amount
    }

    items_lst.append(Items)
    print("Expense added successfully ")
    print("_" * 50)
    # print(items_lst)
```

```python
def view_expenses():
    # print(items_lst)

    for items in items_lst:
        item_name = items['Item']
        item_amount = items['Amount']

        print(f"{item_name} - {item_amount}")
    print("_" * 50)

```

```python
def calculate_total():
    total = 0
    for amount in items_lst:
        total += amount['Amount']

    print(f"Total spent: {total}")
    print("_" * 50)
```

```python
def main():
    while True:
        print("\n1. Add Expense")
        print("2. View Expense")
        print("3. Total Expenses")
        print("4. Exit")

        choice = str(input("\nEnter your choice : "))
        if choice == "1":
            name = str(input("Enter the Item name : "))
            amount = int(input("Enter the Item amount : "))
            add_expenses(name , amount)

        elif choice == "2":
            view_expenses()
        elif choice == "3":
            calculate_total()
        elif choice == "4":
            break
        else:
            print("Invalid Choice")
main()
```
