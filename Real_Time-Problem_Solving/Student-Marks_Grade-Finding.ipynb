```python

# STUDENT MARKS PROBLEM BY USING (LIST,DICTIONARY ,TUPLE , SET)

"""Students: Raj, Priya, Amit, Sneha, Karan

   Marks:
   Raj: Math=85, Science=78, English=90
   Priya: Math=92, Science=88, English=85
   Amit: Math=70, Science=75, English=68
   Sneha: Math=95, Science=90, English=92
   Karan: Math=80, Science=82, English=78
"""


# OUTPUT SHOULD BE LIKE

"""All Students: ['Raj', 'Priya', 'Amit', 'Sneha', 'Karan']
  First 3 Students: ['Raj', 'Priya', 'Amit']

  Raj - Average: 84.33 - Grade: B
  Priya - Average: 88.33 - Grade: A
  Amit - Average: 71.00 - Grade: B
  Sneha - Average: 92.33 - Grade: A
  Karan - Average: 80.00 - Grade: B

  Top Student: Sneha
  Unique Grades: {'A', 'B'}
"""
```

```python

# storing sduents data in list form
Students = ['Raj', 'Priya', 'Amit', 'Sneha', 'Karan']

Marks = {

    "S1" : {
        "name" : "Raj",
        "marks":{
            "Math":85,
            "Science":78,
            "English":90
        }
    },

    "S2" : {
        "name" : "Priya",
        "marks":{
            "Math":92,
            "Science":88,
            "English":85
        }
    },

    "S3" : {
        "name" : "Amit",
        "marks":{
            "Math":70,
            "Science":75,
            "English":68
        }
    },

    "S4" : {
        "name" : "Sneha",
        "marks":{
            "Math":95,
            "Science":90,
            "English":92
        }
    },

    "S5" : {
        "name" : "Karan",
        "marks":{
            "Math":80,
            "Science":82,
            "English":78
        }
    },

}

print(f"All Students : {Students}")
first_3 = Students[0:3]         # Printing first 3 names in the list
print(f"First 3 Students : {first_3}\n")

top_scorer = 0
top_name = ""
unq = set()

for student ,details in Marks.items():

    Average = sum(details["marks"].values()) / len(details["marks"].values())
    if Average >= 85:
        grade = "A"
    elif Average >=70:
        grade="B"
    else :
        grade= "C"
    print(f"{details["name"]} - Average : {Average:.2f} - Grade : {grade}")


    # Top Scorer from the above average
    if Average > top_scorer:
        top_scorer = Average
        top_name = details["name"]

    # Unique grades to store in the set
    unq.add(grade)
    # print(unq)

print(f"\nTop Student: {top_name}")
print("Unique Grades : ",unq)

```
