```python

"""
1. Add Book
2. View Books
3. Borrow Book
4. Exit

Enter choice: 1
Enter book name: Python Basics
Enter copies: 3
Book added successfully



Borrow Book : -

Enter book name: Python Basics
Book borrowed successfully
Remaining copies: 2

"""
```

```python

book_list = []

def add_book(name , copies):
    items_list = {
        "book" : name,
        "copies" : copies
    }

    book_list.append(items_list)
    print("Book Added Succesfully")
    print(book_list)
    print("_" * 45)

```

```python

def view_book():
    for view in book_list:
        book_view = view["book"]
        copies_view = view["copies"]
        print(f"{book_view} - {copies_view}")
    print("_" * 45)
```

```python

def borrow_book(borrow):
    borrow_book_name = ""
    remaining_copies = 0
    found = False

    for borrow_book in book_list:
        # print(borrow_book)
        if borrow.lower() == borrow_book["book"]:
            borrow_book_name = borrow_book["book"]
            remaining_copies = borrow_book["copies"] - 1
            print(borrow_book_name)
            print("Book Borrowed Successfully")

            found = True

            if remaining_copies > 0 :
                print(f"Remaining Copies of {borrow_book_name} are {remaining_copies}")
            else:
                print("Book Not Available")

    if not found:
        print("Book cannot found")

# borrow_book("Social")
```

```python

def main():
    while True:
        print("\n1. Add Book")
        print("2. View Books")
        print("3. Borrow Book")
        print("4. Exit\n")

        choice = str(input("Enter the choice : "))

        if choice == "1":
            name = str(input("Enter the Book name : "))
            copies = int(input("Enter the Copies : "))
            add_book(name , copies)

        elif choice == "2":
            view_book()
        elif choice == "3":
            borrow = str(input("Enter the book name : "))
            borrow_book(borrow)
        elif choice == "4":
            break
main()
```
