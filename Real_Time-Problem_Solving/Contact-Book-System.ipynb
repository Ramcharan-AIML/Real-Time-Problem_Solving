```python
# Output should be like this

"""

1. Add Contact
2. View Contacts
3. Search Contact
4. Exit

Enter choice: 1
Enter name: Ravi
Enter phone: 9876543210
Contact added successfully

"""
```

```python
contact_lst = []

def add_contact(name , phno):

    contact_details = {
        "cnct_name" : name,
        "mobile_no" : phno
    }

    contact_lst.append(contact_details)
    print("Contact Added Successfully")
    print("_"*45)
    # print(contact_lst)
```

```python
def view_contact():
    # print(contact_lst)
    for name in contact_lst:
        cnct_name = name["cnct_name"]
        mobile_no = name["mobile_no"]

        print(f"{cnct_name} - {mobile_no}")
    print("_"*45)
```

```python
def search_contact(search_name):

    viewname = ""
    viewname_mob = 0

    found = False

    for s_name in contact_lst:
        # print(s_name)
        # print(s_name["cnct_name"] , s_name["mobile_no"])
        # print(s_name)

        if search_name in s_name["cnct_name"]:
            viewname = s_name["cnct_name"]
            viewname_mob = s_name["mobile_no"]
            print(f"{viewname}-{viewname_mob}")
            print("_"*45)
            found = True
    
    if not found:
        print("Contact not found")
# search_contact("sunil")
```

```python
def main():
    while True:
        print("\n1. Add Contact")
        print("2. View Contacts")
        print("3. Search Contact")
        print("4. Exit")
        

        choice = str(input("\nEnter the chocie : "))
        # print("_" * 45)

        if choice == "1":
            name = str(input("Enter the name : "))
            phno = int(input("Enter the Phone number : "))
            add_contact(name , phno)

        elif choice == "2":
            view_contact()
        elif choice == "3":
            search_name = str(input("Enter name to find Number : "))
            search_contact(search_name)
        elif choice == "4":
            break

main()
```
