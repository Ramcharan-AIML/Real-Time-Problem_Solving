```python

def welcome():
    print("=" * 50)
    print("I am ATM Machineüèßüèß , How can i help you today")
    print("=" * 50)
```

```python
# @title
def main_code():
    account_name = "ram"
    PIN = 1234
    balance = 10000

    welcome()           #calling function Welcome to ATM

    name = str(input("Enter the Your Name : "))

    valid = False

    if name == account_name:
        print(f"\nWelcome {account_name}")
        pin_code = int(input("Enter the your PIN : "))
        if pin_code == PIN:
            valid = True
            print("\nChoose your option to Proceed")
            print("üëâ1)Check Balance")
            print("üëâ2)Withdrawal Money")
            print("üëâ3)Exit \n")

            while valid:

                choice = int(input("Enter Your Choice :"))

                if choice == 1:
                    print(f"Your Balance is {balance}rs\n")
                elif choice == 2:
                    amount = int(input("Enter amount to withdraw ü§ëü§ë:"))
                    if amount <= balance:
                            print("Withdraw Successfullüí∏üí∏")
                            balance = balance - amount
                            print(f"Remaining balance is : {balance} \n")
                    else:
                        print("‚ùåInsufficient Balance")

                elif choice == 3:
                    print("Thank you for choosing ATM")
                    break
                else:
                    print("‚ùåIvalid Choice")


        else:
            print("‚ùåPlease check PIN")
    else:
        print("‚ùåunauthorised user")    #vefifying the user first  / calling function

main_code()
```
